```python
#1.who is the strongest pokemon and weakest
print("strongest pokemon is----",df[df["Total"]==df["Total"].max()][["Name","Total"]])
print("weakest pokemon is----",df[df["Total"]==df["Total"].min()][["Name","Total"]])
```

```python
#2. find out the characteristics of pikachu pokemon
df.set_index(df["Name"],inplace=True)
for i in df["Name"]:
    if i=="Pikachu":
        print(df.loc["Pikachu",:])   
df[df["Name"]=="Pikachu"]
```

```python
#3. how many legends pokemon
print(df["Legendary"].unique())
print(df["Legendary"].value_counts())
```

```python
#4. up to which generation pokemon exists in the catalog
df["Generation"].unique()
```

```python
# which generartion has more pokemon
df["Generation"].value_counts()
```

```python
#5. which category of pokemon is strongest and weakest
print("strongest category of pokemon is----",df[df["Total"]==df["Total"].max()][["Type 1","Total"]])
print("weakest category of pokemon is----",df[df["Total"]==df["Total"].min()][["Type 1","Total"]])
```

```python
#6. find top 5 strongest pokemon based on total
df["Total"].sort_values(ascending=False).head()
```

```python
#7. find max, min, and avg stats(total) of each category(type) of pokemon
df.groupby("Type 1")["Total"].agg(["mean","min","max"])
```

```python
#8. which pokemon whose total point is max and min(name of pokemon)
print(df[df["Total"]==df["Total"].max()])
print(df[df["Total"]==df["Total"].min()])
```

```python
df.drop(columns=["Name"],inplace=True)
df.reset_index(inplace=True)
```

```python
#9. find avg, max, and min of total stats of each type of pokemon along with pokemon names
df.groupby(["Type 1","Name"])["Total"].agg(["mean","min","max"])
```

```python
# 10. which generation' pokemon is strongest
df.groupby(["Generation"])["Total"].agg(["mean","min","max"]).sort_values("max",ascending=False)
```

```python
#11. find top five strongest pokemon based on hit point(HP) use sort_values()
df.groupby(["Name","Type 1"])["HP"].max().sort_values(ascending=False).head()
```

```python
#12. find max, min, and avg hit point of each type
df.groupby(["Type 1"])["HP"].agg(["mean","min","max"]).sort_values("max",ascending=False)
```
